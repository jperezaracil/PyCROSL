```python
import numpy as np
import pandas as pd
import xlsxwriter
```

```python
data_dyn1 = pd.read_csv("achived/cro_dyn_ex1.csv", index_col=0)
data_dyn2 = pd.read_csv("achived/cro_dyn_ex2.csv", index_col=0)
data_dyn3 = pd.read_csv("achived/cro_dyn_ex3.csv", index_col=0)
data_dyn = pd.concat([data_dyn1, data_dyn2, data_dyn3])
data_dyn
```

```python
data_nodyn1 = pd.read_csv("achived/cro_nodyn_ex1.csv", index_col=0)
data_nodyn2 = pd.read_csv("achived/cro_nodyn_ex2.csv", index_col=0)
data_nodyn = pd.concat([data_nodyn1, data_nodyn2])
data_nodyn
```

```python
data_classic1 = pd.read_csv("achived/cro_classic_ex1.csv", index_col=0)
data_classic2 = pd.read_csv("achived/cro_classic_ex2.csv", index_col=0)
data_classic3 = pd.read_csv("achived/cro_classic_ex3.csv", index_col=0)
data_classic = pd.concat([data_classic1, data_classic2, data_classic3])
data_classic
```

```python
data_dyn_grouped = data_dyn.groupby(data_dyn["substrates"])
data_nodyn_grouped = data_nodyn.groupby(data_nodyn["substrates"])
data_classic_grouped = data_classic.groupby(data_classic["substrates"])
```

```python
#writer = pd.ExcelWriter("CRO_SL_python.xlsx", engine="xlsxwriter")
with pd.ExcelWriter("CRO_SL_python.xlsx", engine="xlsxwriter") as writer:
    for i in pd.unique(data_dyn["substrates"]):
        table_dyn = data_dyn_grouped.get_group(i)
        table_dyn.pop("substrates")

        table_nodyn = data_nodyn_grouped.get_group(i)
        table_nodyn.pop("substrates")
        table_nodyn.pop("function")

        table_classic = data_classic_grouped.get_group(i)
        table_classic.pop("substrates")
        table_classic.pop("function")

        s_name = f"{i[:15]}(30); eval=3e5"
        table_dyn.to_excel(writer, sheet_name=s_name, index=False, startrow=1)
        table_nodyn.to_excel(writer, sheet_name=s_name, index=False, startrow=1, startcol=4)
        table_classic.to_excel(writer, sheet_name=s_name, index=False, startrow=1, startcol=7)
        writer.sheets[s_name].merge_range("B1:D1", "CRO_SL-Dyn")
        writer.sheets[s_name].merge_range("E1:G1", "CRO_SL")
        writer.sheets[s_name].merge_range("H1:J1", "CRO_SL-Classic")
```
